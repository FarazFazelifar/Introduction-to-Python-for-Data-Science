```python
import pandas as pd
import numpy as np

## read csv
df = pd.read_csv('data/pokemon_data.csv')
df.head(5)
```

```python
df.describe()
```

```python
df.groupby(['Type 1']).count()
```

```python
df.drop_duplicates(inplace=True)
```

```python
numeric_cols = df.select_dtypes(include=[np.number]).columns
for col in numeric_cols:
    df[col].fillna(df[col].mean(), inplace=True)

object_cols = df.select_dtypes(exclude=[np.number]).columns
for col in object_cols:
    df[col].fillna(value="", inplace=True)
```

```python
df.describe()
```

```python
df.groupby(['Type 1']).count()
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Example: 3D Bar Plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
# Sample data for 3D bars
_x = np.arange(4)
_y = np.arange(4)
_xx, _yy = np.meshgrid(_x, _y)
x, y = _xx.ravel(), _yy.ravel()
z = np.zeros_like(x)
dx = dy = 0.5 * np.ones_like(z)
dz = np.random.rand(len(z))
ax.bar3d(x, y, z, dx, dy, dz, color="teal", shade=True)
ax.set_title("3D Bar Plot: Volumetric Data")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.set_zlabel("Value")
plt.show()
```

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.sin(x) + 0.5
plt.figure(figsize=(10, 6))
plt.plot(x, y1, label='sin(x)', color="blue")
plt.plot(x, y2, label='sin(x)+0.5', color="red")
plt.fill_between(x, y1, y2, color='gray', alpha=0.3)
# 'fill_between' shades the area between y1 and y2 with transparency 'alpha'
plt.title("Fill Between: Area Between Two Curves")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

```python
for type_ in df['Type 1'].unique():
    type_data = df[df['Type 1'] == type_]
    Q1 = type_data['Attack'].quantile(0.25)
    Q3 = type_data['Attack'].quantile(0.75)
    IQR = Q3 - Q1
    outliers = type_data[
        (type_data['Attack'] < (Q1 - 1.5 * IQR)) | 
        (type_data['Attack'] > (Q3 + 1.5 * IQR))
    ]
    if len(outliers) > 0:
        print(f"\nOutliers for {type_}:")
        print(outliers[['Name', 'Attack', 'Legendary']])
```
